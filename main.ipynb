```python
import numpy as np
import pandas as pd
import datetime as dt
```

# AGREGAR DATOS DESDE UN EXCEL

```python
df = pd.read_csv('./Data/dataframe.csv')

df.tail()
```

```python
sheet = 'Volador'

recimed = pd.read_excel('./Data/recimed2023.xlsx', sheet_name=sheet, usecols='A:C', skiprows=6, nrows=14)

recimed
```

```python
recimed['MATERIAL']
```

```python
year = '2023'

calendar = {
    'ENERO': '01',
    'FEBRERO ': '02'
    #'MARZO': '03',
    #'ABRIL': '04',
    #'MAYO': '05',
    #'JUNIO': '06',
    #'JULIO': '07',
    #'AGOSTO': '08',
    #'SEPTIEMBRE': '09',
    #'OCTUBRE': '10',
    #'NOVIEMBRE': '11',
    #'DICIEMBRE': '12'
}
```

```python
recimeDict = {
    'Fecha': list(),
    'Nombre_Material': list(),
    'Cantidad_kg': list(),
    'Nombre_Categoria': list(),
    'Operador_logistíco': list(),
    'Sede': list(),
    'YEAR': list(),
    'MONTH': list()
}

zero = 0
for i in range(0, recimed['MATERIAL'].count(), 1):
    for k,v in calendar.items():
        # FECHA
        recimeDict['Fecha'].append(year + '-' + v + '-28')
        recimeDict['MONTH'].append(k[:3:].capitalize())
        # MATERIAL
        if recimed['MATERIAL'][i] == 'PLEGADIZA':
            material = 'Plega'
            recimeDict['Nombre_Material'].append(material)
        elif (recimed['MATERIAL'][i] == 'PET') or (recimed['MATERIAL'][i] == 'PVC'):
            material = recimed['MATERIAL'][i]
            recimeDict['Nombre_Material'].append(material)
        elif (recimed['MATERIAL'][i].lower() == 'cubetas de huevo') or (recimed['MATERIAL'][i].lower() == 'vaso desechable'):
            material = recimed['MATERIAL'][i].lower().split(' ')
            material[0] = material[0].capitalize()
            recimeDict['Nombre_Material'].append(' '.join(material))
        elif recimed['MATERIAL'][i] == 'TUBERIA PVC':
            material = 'PVC'
            recimeDict['Nombre_Material'].append(material)
        else:
            material = recimed['MATERIAL'][i].capitalize()
            recimeDict['Nombre_Material'].append(material)
        # CATEGORIA DEL MATERIAL
        if material in ['Acero', 'Aluminio Karla','Aluminio Sucio','Aluminio Limpio','Aluminio Perciana', 'Cobre Rojo', 'Cobre Amarillo', 'Cobre Encauchetado','Chatarra']:
            recimeDict['Nombre_Categoria'].append('Metal')
        elif material in ['Vidrio']:
            recimeDict['Nombre_Categoria'].append('Vidrio')
        elif material in ['PET', 'Plástico', 'PVC','Pasta','Vaso desechable']:
            recimeDict['Nombre_Categoria'].append('Plástico')
        elif material in ['Plega', 'Tetrapack','Periódico','Catón','Archivo','Cubetas de huevo']:
            recimeDict['Nombre_Categoria'].append('Papeles y Cartones')
        else:
            recimeDict['Nombre_Categoria'].append('Otros')
        # CANTIDAD EN KG
        if str(recimed.iloc[i][k]) != 'nan':
            recimeDict['Cantidad_kg'].append(recimed.iloc[i][k])
        else:
            recimeDict['Cantidad_kg'].append(0)
        # ZERO
        zero += 1

recimeDict['Operador_logistíco'] = ['Recimed' for i in range(0, zero, 1)]
recimeDict['Sede'] = [sheet for i in range(0, zero, 1)]
recimeDict['YEAR'] = [year for i in range(0, zero, 1)]

recimeDF = pd.DataFrame(recimeDict)
recimeDF.tail()
```

```python
for i in range(0, zero, 1):
    df.loc[i + df['Fecha'].count()] = recimeDF.iloc[i]

df.tail()
```

# SUBESTACIONES

```python
sub = pd.read_csv('Data/subestaciones.csv')

sub.pop('Operador_logistíco')

sub.head()
```

```python
calendar = {
    '03': 'Mar',
    '04': 'Abr',
    '05': 'May',
    '06': 'Jun',
    '07': 'Jul',
    '08': 'Ago',
    '09': 'Sept',
    '10': 'Oct',
    '11': 'Nov'
}
```

```python
sub['Cantidad_kg'] = [float(str(i).replace(',', '.')) for i in list(sub['Cantidad_kg'])]
sub['Operador_logistíco'] = [np.nan for i in range(0, sub['Fecha'].count(), 1)]
sub['Nombre_Categoria'] = [np.nan for i in range(0, sub['Fecha'].count(), 1)]
sub['YEAR'] = ['2022' for i in range(0, sub['Fecha'].count(), 1)]
sub['MONTH'] = [calendar[i.split('-')[1]] for i in list(sub['Fecha'])]

sub.head()
```

```python
df = pd.read_csv('Data/dataframe.csv')

df.tail()
```

```python
for i in range(0, sub['Fecha'].count(), 1):
    df.loc[i + df['Fecha'].count()] = sub.iloc[i]

df.tail()
```
